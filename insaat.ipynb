```python
import torch
from segmentation_models_pytorch import UnetInfluence
```

```python
clsf_aux_params = dict(pooling='avg',            # one of 'avg', 'max'
                        dropout=None, #0.5         # dropout ratio, default is None
                        activation='sigmoid',      # activation function, default is None
                        classes=1,                 # define number of output labels
                        )

model = UnetInfluence(in_channels= 14,
                                patch_size=256,
                                batch_size=8,
                                classes=1,
                                occlusion_modality=6,
                                aux_params=clsf_aux_params)
```

```python
model.eval()
```
